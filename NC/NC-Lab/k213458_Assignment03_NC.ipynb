```python
import matplotlib.pyplot as plt
import numpy as np
# Step 1: Generate data

x = np.random.rand(10, 1)
y = 2 * x + np.random.randn(10, 1)
print(x)
print("\n")
print(y)
# Step 2: Define cost function
def J(theta_0, theta_1, x, y):
    m = len(x)
    h = theta_0 + theta_1 * x
    return 1 / (2 * m) * np.sum((h - y) ** 2)

# Step 3: Implement gradient descent
alpha = 0.1
theta_0 = 0
theta_1 = 0
iterations = 100
J_history = np.zeros(iterations)

for i in range(iterations):
    h = theta_0 + theta_1 * x
    J_history[i] = J(theta_0, theta_1, x, y)
    temp0 = theta_0 - alpha * (1/len(x)) * np.sum(h - y)
    temp1 = theta_1 - alpha * (1/len(x)) * np.sum((h - y) * x)
    theta_0 = temp0
    theta_1 = temp1
    print ("theta0 {}, theta1 {}, J(theta0, theta1) {} iteration {}".format(theta_0,theta_1,J_history[i], i))
# Step 4: Plot J(theta_0, theta_1)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

theta_0_vals = np.linspace(-2, 2, 100)
theta_1_vals = np.linspace(-2, 2, 100)
J_vals = np.zeros((len(theta_0_vals), len(theta_1_vals)))

for i, theta_0 in enumerate(theta_0_vals):
    for j, theta_1 in enumerate(theta_1_vals):
        J_vals[i, j] = J(theta_0, theta_1, x, y)

theta_0_vals, theta_1_vals = np.meshgrid(theta_0_vals, theta_1_vals)
ax.plot_surface(theta_0_vals, theta_1_vals, J_vals, cmap='viridis')
ax.set_xlabel('Theta 0')
ax.set_ylabel('Theta 1')
ax.set_zlabel('J')

plt.show()

# Plot point for initial theta_0, theta_1
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(theta_0_vals, theta_1_vals, J_vals, cmap='viridis')
ax.set_xlabel('Theta 0')
ax.set_ylabel('Theta 1')
ax.set_zlabel('J')
ax.scatter(theta_0, theta_1, J_history[0], c='r')

plt.show()
```
